```python
import torch
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("./IMG/0998.png")
plt.imshow(img[:, :, ::-1])
plt.show()
```

```python
sr = cv2.dnn_superres.DnnSuperResImpl_create()
```

```python
path = './EDSR/models/EDSR_x4.pb' # About 3min
```

## CV_EDSR result (without Kernel Estimation)

```python
sr.readModel(path)
sr.setModel("edsr", 4)

result = sr.upsample(img)

plt.figure(figsize=(12, 8))

plt.subplot(1, 2, 1)
plt.imshow(img[:, :, ::-1])

plt.subplot(1, 2, 2)
plt.imshow(result[:, :, ::-1])

plt.show()
```
